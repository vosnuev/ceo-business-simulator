## EDA 용 간단한 전처리

```python
# 데이터 타입 정리
datetime_cols = [
    'user_creation_datetime',
    'confirmation_datetime',
    'first_purchase_datetime',
    'last_purchase_datetime',
    'first_integration_datetime',
    'last_segment_recorded'
]

for col in datetime_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

df[datetime_cols].dtypes
```

```python
# 데이터 타입별 컬럼 분류
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("범주형 변수 개수:", len(categorical_cols))
print("수치형 변수 개수:", len(numerical_cols))

print("\n범주형 변수:")
print(categorical_cols)

print("\n수치형 변수:")
print(numerical_cols)
```

## 변수 시각화

```python
# 타겟 변수 시각화
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution (0=ok, 1=churned)')
plt.show()
```

```python
# 범주형 변수별 이탈률 확인 (user_id, status 제외)
cat_cols_to_check = ['platform', 'market_country']

for col in cat_cols_to_check:
    churn_rate = df.groupby(col)['Churn'].mean().sort_values(ascending=False)
    churn_rate.plot(kind='bar')
    plt.title(f'{col}별 이탈률')
    plt.ylabel('Churn Rate')
    plt.show()
```

- 뭔지 몰라서 의미 없음

```python
# 수치형 변수 요약 통계 및 왜도 계산
summary = df[numerical_cols].describe().T
summary['skew'] = df[numerical_cols].skew()
summary['IQR'] = summary['75%'] - summary['25%']

summary = summary[['mean', '50%', 'std', 'min', '25%', '75%', 'max', 'IQR', 'skew']]
summary.sort_values(by='skew', ascending=False)
```

```python
core_cols = [
    'total_sales', 'frequency', 'recency',
    'historic_revenue_eur', 'historic_aov'
]

for col in core_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'{col} 분포', fontsize=14, fontweight='bold')
    plt.show()
```

```python
for col in core_cols:
    sns.histplot(np.log1p(df[col]), kde=True)
    plt.title(f'{col} 로그 분포', fontsize=14, fontweight='bold')
    plt.show()
```

- 웨도 심함!!

```python
import math

cols = [
    'total_sales', 'frequency', 'recency',
    'historic_revenue_eur', 'historic_aov',
    'avg_sales_per_day', 'longevity'
]

n_cols = 3  # 한 줄에 그래프 개수
n_rows = math.ceil(len(cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))

axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col, fontsize=12)

# 남는 칸 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('수치형 변수 이상치 한눈에 보기', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 이탈 VS 유지 비교

```python
# 평균 중앙값 비교
mean_df = df.groupby('Churn')[core_cols].mean().T
median_df = df.groupby('Churn')[core_cols].median().T

mean_df.columns = ['유지', '이탈']
median_df.columns = ['유지(중앙값)', '이탈(중앙값)']

pd.concat([mean_df, median_df], axis=1)
```

```python
for col in core_cols:
    plt.figure(figsize=(9, 5))
    sns.boxplot(data=df, x='Churn', y=col)
    plt.title(f'{col} - 이탈 여부 비교', fontsize=14, fontweight='bold')
    plt.xlabel('Churn (0=유지, 1=이탈)')
    plt.show()
```

```python
# KDE
for col in core_cols:
    plt.figure(figsize=(10, 5))
    sns.kdeplot(df[df['Churn']==0][col], label='유지', fill=True)
    sns.kdeplot(df[df['Churn']==1][col], label='이탈', fill=True)
    plt.title(f'{col} 분포 비교', fontsize=14, fontweight='bold')
    plt.legend()
    plt.show()
```

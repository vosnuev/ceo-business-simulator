비교적 단순한 고객 속성 기반 이탈 데이터(ecommerce_customer_data)를 사용한 베이스라인 모델 학습 노트북  
  
1) 데이터 개요  
  
이 데이터는 총 2000행, 9개 컬럼으로 구성되어 있으며, 각 행은 고객 1명을 의미한다.  
타깃 변수는 Churn이며, 고객의 기본 속성과 소비 행동을 이용하여 이탈 여부를 예측하는 구조이다.  
  
2) 컬럼 설명  
Customer_ID: 고객 식별자  
Age: 나이  
Gender: 성별  
Annual_Income: 연소득  
Spending_Score: 소비 성향 점수  
Membership_Years: 멤버십 유지 기간  
Online_Purchases: 온라인 구매 횟수  
Discount_Usage: 할인 사용 정도  
Churn: 이탈 여부 (타깃)  
  
3) 전처리 방향  
  
노트북에서는 아래와 같은 전처리 흐름을 사용하였다.
  
Customer_ID 제거  
→ 식별자는 예측에 직접적 의미가 없고, 모델이 의미 없는 번호 패턴을 학습할 수 있기 때문  
Gender 인코딩  
→ 범주형 변수이므로 pd.get_dummies(..., drop_first=True)를 사용하여 숫자형 변수로 변환  
train / validation / test 분할  
→ 데이터 누수를 방지하고 모델 선택 및 최종 평가를 분리하기 위함  
→ stratify=y로 클래스 비율 유지  
StandardScaler 적용  
→ Logistic Regression은 입력 스케일에 민감하므로 train 데이터 기준으로 표준화  
→ Decision Tree, Random Forest는 스케일링 없이 원본 데이터 사용 가능  

4)  EDA 관점 해석  
  
이 데이터는 전처리 난이도가 매우 낮은 편이며, 머신러닝 입문용 tabular classification 데이터로 적합하다.  
다만 수치형 변수와 타깃 간의 단순 선형 관계가 아주 강한 편은 아니라서, 단순 선형 모델(Logistic Regression)보다 트리 계열 모델(Random Forest, XGBoost 등) 이 더 유리할 가능성이 있다.  

5) 노트북에서 사용한 모델 선정 이유와 결과 해석  
  
노트북 1에서는 전자상거래 고객 이탈 예측 문제를 해결하기 위해 다음과 같은 머신러닝 모델을 단계적으로 적용하였다.  
  
Logistic Regression  
Decision Tree  
Random Forest  
VotingClassifier  
XGBoost  

이 모델들은 단순한 선형 모델부터 비선형 관계와 변수 간 상호작용을 학습할 수 있는 트리 기반 앙상블 모델까지 점진적으로 확장하는 구조로 선택되었다. 즉, 먼저 해석이 쉬운 baseline 모델로 출발한 뒤, 데이터의 복잡한 패턴을 더 잘 반영할 수 있는 고성능 모델로 발전시키는 흐름을 따랐다고 볼 수 있다.  

5.1) Logistic Regression을 선택한 이유  
  
Logistic Regression은 이진 분류 문제에서 가장 기본적인 기준선(baseline) 모델이다.  
이 모델은 입력 변수와 타깃 사이의 관계를 선형적으로 가정하며, 구조가 단순하고 학습 속도가 빠르며 해석이 용이하다는 장점이 있다.  
  
노트북 1 데이터는 고객의 나이, 연소득, 소비 점수, 온라인 구매 횟수, 할인 사용 정도 등 비교적 정형화된 표 데이터이기 때문에, 가장 먼저 Logistic Regression을 적용하는 것은 자연스러운 선택이다. 특히 이 모델을 통해 데이터가 단순 선형 관계만으로도 어느 정도 구분 가능한지를 먼저 점검할 수 있다.  
  
그러나 실제 결과를 보면 Logistic Regression은 높은 정확도를 보여주지 못했고, 특히 이탈 고객(클래스 1)을 충분히 잡아내지 못하는 한계를 보였다. 이는 데이터의 관계가 단순한 선형 구조라기보다, 비선형 패턴과 상호작용이 일부 포함되어 있음을 시사한다.  
  
5.2) Decision Tree를 선택한 이유  
  
Decision Tree는 입력 변수의 값을 기준으로 규칙을 나누어 예측하는 모델로, 선형 모델보다 비선형 관계를 더 잘 반영할 수 있다는 장점이 있다. 또한 “어떤 조건에서 고객이 이탈하는가”를 규칙 형태로 확인할 수 있어 해석도 비교적 쉽다.  
  
노트북 1 데이터에서는 단순 선형 상관이 아주 강하지 않았기 때문에, Logistic Regression 다음 단계로 Decision Tree를 사용하는 것은 타당하다. 즉, 고객 이탈이 단순 직선 경계로 나뉘는 것이 아니라 특정 구간 조건이나 조합 규칙으로 나뉠 가능성을 확인하기 위한 선택이었다.  
  
실제 결과에서는 Logistic Regression보다 약간 더 복잡한 패턴을 학습했지만, 여전히 validation 성능이 충분히 높지 않았고, class 1 예측 성능도 제한적이었다. 이는 단일 트리가 데이터의 복잡성을 어느 정도 잡아내기는 했지만, 일반화 성능 측면에서는 부족했음을 의미한다.  
  
5.3) Random Forest를 선택한 이유  
  
Random Forest는 여러 개의 결정트리를 독립적으로 학습시킨 뒤 결과를 종합하는 앙상블 모델이다. 단일 Decision Tree보다 과적합을 줄이고, 더 안정적인 성능을 보이는 것이 일반적이다. 또한 비선형 관계와 feature interaction을 잘 다룰 수 있어, tabular 데이터에서 매우 자주 사용된다.  
  
노트북 1 데이터는 각 변수 간 단순 선형 상관이 강하지 않고, 고객 특성들 간 복합적인 조합이 churn에 영향을 미칠 가능성이 있다. 따라서 Random Forest는 이러한 비선형 구조와 변수 간 상호작용을 포착하기 위한 모델로 적절하게 선택되었다.  
  
실험 결과 Random Forest는 Logistic Regression과 Decision Tree보다 더 나은 가능성을 보여주었으나, 초기에 상당한 과적합 양상이 나타났다. train 성능은 높았지만 validation 성능이 충분히 따라오지 못했으며, 특히 class 1 탐지 성능이 안정적이지 못했다. 이후 하이퍼파라미터를 조정하면서 과적합을 완화했지만, validation 기준 최종 성능은 여전히 제한적이었다.  
  
5.4) VotingClassifier를 시도한 이유  
  
VotingClassifier는 여러 모델의 예측을 합쳐 최종 결정을 내리는 앙상블 기법이다. 노트북 1에서는 Logistic Regression, Decision Tree, Random Forest를 결합하여 성능을 개선하려는 시도가 이루어졌다.  
  
이 방법을 사용한 이유는 각 모델이 학습하는 패턴의 성격이 다르기 때문이다.  
  
Logistic Regression: 선형 관계 중심  
Decision Tree: 규칙 기반 분기  
Random Forest: 다수 트리 기반 비선형 앙상블  
  
즉 서로 다른 관점을 가진 모델을 결합하면, 단일 모델보다 더 안정적이고 강한 성능을 기대할 수 있다.  
  
그러나 실제 실험에서는 VotingClassifier가 기대만큼 좋은 결과를 보이지 않았다. threshold 조정에 따라 결과가 극단적으로 흔들리거나, 특정 클래스만 거의 전부 예측하는 문제가 나타났다. 이는 개별 모델들이 충분히 강한 상태가 아니었고, 특히 클래스 불균형과 threshold 설정 문제까지 겹치면서 soft voting이 안정적인 개선으로 이어지지 못했기 때문으로 해석할 수 있다.  
  
5.5) 최종적으로 XGBoost를 선택한 이유  
  
XGBoost는 Gradient Boosting 기반의 트리 앙상블 모델로, 이전 트리의 오차를 다음 트리가 순차적으로 보완해나가는 구조를 가진다. 표 데이터(tabular data)에서 매우 강력한 성능을 보이는 대표적인 모델이며, 비선형 관계와 변수 간 상호작용을 잘 포착하는 것이 특징이다.  
  
노트북 1에서 최종적으로 XGBoost를 선택한 이유는 다음과 같다.  
  
1) 기존 모델들보다 더 복잡한 패턴을 학습할 가능성이 높았기 때문  
  
Logistic Regression은 선형 가정의 한계가 있었고, Decision Tree는 단일 모델이라 불안정했으며, Random Forest는 과적합을 줄이면서도 성능을 충분히 끌어올리는 데 한계가 있었다. XGBoost는 이들보다 더 정교하게 오차를 보완하면서 학습할 수 있기 때문에, 고객 이탈 패턴을 더 잘 잡아낼 가능성이 있었다.  
  
2) 실제 실험 결과가 가장 유망했기 때문

초기 XGBoost는 과적합이 심한 버전과 너무 보수적인 버전이 번갈아 나타났으나, 하이퍼파라미터를 조정한 이후에는 train과 validation 사이의 균형이 이전보다 더 개선되었다. 특히 train에서 class 1을 일정 수준 이상 예측하면서도 validation accuracy가 기존 머신러닝 모델들보다 더 높게 유지되는 구간이 나타났다.

3) threshold 조정을 통해 class 1 탐지 성능을 더 조절할 수 있었기 때문

노트북 1에서 중요한 목표 중 하나는 이탈 고객(1 클래스)을 더 잘 잡는 것이었다. XGBoost는 predict_proba()를 통해 클래스 1 확률을 제공하므로, 기본 threshold 0.5 외에도 0.35, 0.4 등의 threshold를 validation에서 조정해 recall과 precision의 균형을 탐색할 수 있었다. 이는 단순 accuracy 중심의 판단보다 훨씬 실무적인 접근이다.

4) 최종적으로 가장 “개선 가능성이 큰 모델”로 보였기 때문

Random Forest도 괜찮은 후보였지만 validation 성능이 충분히 강하지 않았고, VotingClassifier는 결과가 불안정했다. 반면 XGBoost는 하이퍼파라미터와 threshold를 조정할수록 성능 균형이 조금씩 개선되었고, class 1 예측도 완전히 포기하지 않는 방향을 보여주었다. 즉 완성형이라기보다 가장 유망한 최종 후보 모델이라는 판단으로 선택한 것이다.

6) 종합 해석

노트북 1의 모델 실험은 단순한 baseline에서 출발하여, 점차 더 복잡한 비선형 모델과 앙상블 모델로 확장하는 구조를 따랐다.

Logistic Regression은 baseline으로 적절했지만, 선형 가정의 한계가 있었다.  
Decision Tree는 비선형 규칙을 일부 잡았지만 일반화 성능이 부족했다.  
Random Forest는 더 강한 트리 앙상블이었으나 과적합과 validation 성능 저하 문제가 있었다.  
VotingClassifier는 여러 모델을 결합하려는 시도였으나, 기대만큼 안정적인 개선을 만들지 못했다.  
XGBoost는 가장 유망한 성능을 보여주었고, 하이퍼파라미터 및 threshold 조정을 통해 추가 개선 가능성이 있는 최종 후보 모델로 판단되었다.  

즉 노트북 1의 실험 흐름은 단순히 여러 모델을 나열한 것이 아니라, 데이터가 선형적으로 잘 분리되지 않는다는 점을 확인한 뒤, 더 강한 비선형 트리 기반 모델로 점진적으로 이동한 과정으로 이해할 수 있다.  

그래서 이 데이터의 학습에는 최종적으로 XGBoost 모델이 사용되었다.

6) 노트북 2에서 사용한 모델 선정 이유와 결과 해석  
  
노트북 2에서는 고객 이탈 예측 문제를 해결하기 위해 다음과 같은 모델들을 순차적으로 적용하였다.  
  
Logistic Regression  
Decision Tree  
Random Forest  
VotingClassifier  
XGBoost  
CatBoost  
  
노트북 2의 데이터는 노트북 1보다 훨씬 복잡한 구조를 가진다.  
단순한 고객 속성뿐 아니라, 플랫폼, 국가, 날짜 정보, 구매 빈도, recency, 매출 지표, AOV, 고객지원(CS) 관련 변수 등 다양한 유형의 정보가 함께 포함되어 있다. 이러한 구조는 단순 선형관계보다는 비선형 관계와 변수 간 상호작용(feature interaction) 이 존재할 가능성이 높다. 따라서 모델 선택 역시 baseline 모델에서 시작하여 점진적으로 더 강한 비선형 트리 기반 앙상블 모델로 확장하는 방식으로 진행되었다.  
  
6.1) Logistic Regression을 선택한 이유  
  
Logistic Regression은 이진 분류 문제에서 가장 대표적인 baseline 모델이다.  
노트북 2에서도 먼저 이 모델을 적용한 이유는 다음과 같다.  
  
해석 가능성이 높다.  
어떤 변수들이 이탈에 선형적으로 영향을 주는지 확인할 수 있다.  
기준 성능을 확인하기 좋다.  
더 복잡한 모델을 적용하기 전에, 단순 선형 경계만으로 어느 정도 분리가 가능한지를 먼저 점검할 수 있다.  
전처리 검증용 baseline 역할을 한다.  
범주형 인코딩, 결측치 처리, 스케일링이 정상적으로 적용되었는지 확인하는 출발점이 된다.  
  
실제 결과를 보면 Logistic Regression은 train과 validation 모두 높은 정확도를 보였으며, 기본적인 분류는 잘 수행하였다.  
이는 노트북 2 데이터가 이전 노트북의 단순 synthetic 데이터보다 훨씬 복잡하지만, 그 안에서도 일부 구간에서는 선형적 분리 가능성이 존재함을 보여준다. 다만 이 데이터가 날짜 파생변수, 매출 지표, recency/frequency 같은 복합 행동 특성을 포함하고 있다는 점을 고려하면, 단순 선형 모델만으로 전체 패턴을 완전히 설명하기는 어렵다고 판단할 수 있다.  
  
6.2) Decision Tree를 선택한 이유  
  
Decision Tree는 규칙 기반으로 데이터를 분할하는 비선형 모델이다.  
이 모델을 선택한 이유는 다음과 같다.  
  
비선형 관계를 반영할 수 있다.  
Logistic Regression이 놓칠 수 있는 구간별 규칙을 포착할 수 있다.  
해석 가능성이 비교적 높다.  
특정 변수의 기준값에 따라 고객이 어떤 방향으로 분류되는지 설명하기 쉽다.  
날짜 파생변수나 recency/frequency 같은 변수의 임계값 효과를 반영하기 적합하다.  
  
노트북 2의 결과에서 Decision Tree는 Logistic Regression과 유사한 수준의 성능을 보였으나, 단일 트리 특성상 모델이 불안정해질 수 있고, 데이터의 복잡성을 완전히 포착하기에는 한계가 있다. 즉 규칙 기반 분할이라는 점에서 의미는 있으나, 최종 모델로 선택하기에는 일반화 성능 측면에서 더 강한 모델이 필요하다고 판단할 수 있다.  
  
6.3) Random Forest를 선택한 이유

Random Forest는 여러 개의 결정트리를 독립적으로 학습시킨 뒤 결과를 종합하는 앙상블 모델이다.
노트북 2에서 이 모델을 적용한 이유는 다음과 같다.

Decision Tree보다 일반화 성능이 더 안정적이다.
단일 트리보다 과적합을 줄이고, 보다 견고한 예측을 만들 수 있다.
비선형 구조와 변수 간 상호작용을 잘 반영한다.
recency, frequency, 매출, 티켓 수 등의 복합 조합을 학습하기에 적절하다.
표 데이터(tabular data)에서 강력한 baseline이 된다.

실제 결과에서도 Random Forest는 train에서 높은 성능을 보였고, validation에서도 비교적 안정적인 예측을 수행하였다.
다만 단일 트리보다 성능은 향상되었지만, 부스팅 계열 모델만큼의 세밀한 오차 보정 능력은 부족하였다. 즉 비선형 패턴을 잘 반영하는 유용한 모델이었지만, 보다 강한 예측력이 필요한 상황에서는 한 단계 더 고도화된 모델이 필요했다.

6.4) VotingClassifier를 시도한 이유

VotingClassifier는 여러 모델의 예측을 합쳐 최종 결정을 내리는 앙상블 기법이다.
노트북 2에서는 Logistic Regression, Decision Tree, Random Forest를 결합하여 soft voting을 수행하였다.

이 방식을 선택한 이유는 다음과 같다.

서로 다른 성격의 모델을 결합하면 안정성이 높아질 수 있다.
Logistic Regression: 선형 패턴
Decision Tree: 규칙 기반 비선형 패턴
Random Forest: 비선형 앙상블 패턴
단일 모델의 약점을 보완할 수 있다.
예를 들어 한 모델이 놓친 샘플을 다른 모델이 잡아줄 가능성이 있다.
복잡한 tabular 데이터에서 앙상블의 효과를 시험해볼 수 있다.

실험 결과 VotingClassifier는 validation/test에서 비교적 좋은 성능을 보였으며, 개별 모델보다 조금 더 안정적인 결과를 내는 구간도 확인되었다.
그러나 Voting은 어디까지나 서로 다른 모델의 평균적 결정을 활용하는 방식이므로, 데이터의 복잡한 구조를 직접적으로 학습하는 고성능 부스팅 계열 모델에 비해서는 한계가 있다. 즉 유의미한 중간 단계 실험이었지만, 최종 선택 모델로는 더 강한 표현력을 가진 모델을 검토할 필요가 있었다.

6.5) XGBoost를 선택한 이유

XGBoost는 Gradient Boosting 기반 트리 앙상블 모델로, 이전 트리가 틀린 부분을 다음 트리가 순차적으로 보완해 나가는 구조를 가진다.
노트북 2에서 XGBoost를 선택한 이유는 다음과 같다.

1) 노트북 2 데이터의 특성과 잘 맞기 때문이다

이 데이터는 단순 선형 구조라기보다 다음과 같은 복합 구조를 포함한다.

비선형 관계
feature interaction
날짜에서 파생된 기간 차이 정보
recency / frequency / AOV / 매출 변수의 복합적 패턴
결측치 처리 후 남은 수치형·범주형 정보의 혼합 구조

이러한 데이터는 선형 모델보다 트리 기반 boosting 모델이 더 잘 다룰 가능성이 높다. XGBoost는 특히 tabular 데이터에서 매우 강력한 성능을 보이는 대표적인 모델이다.

2) 실제 성능이 가장 강력한 후보로 나타났기 때문이다

노트북 2의 실험 결과를 보면, XGBoost는 train 성능이 매우 높았고 validation에서도 높은 accuracy를 유지하였다.
특히 validation 기준으로:

accuracy가 약 0.95 수준까지 올라갔고
confusion matrix에서도 0과 1 클래스를 모두 상당히 잘 구분하는 모습을 보였다

이는 Logistic Regression, Decision Tree, Random Forest보다 한 단계 높은 성능을 보여주는 결과로 해석할 수 있다.

3) 비선형성과 변수 중요도를 함께 확인할 수 있기 때문이다

XGBoost는 단순히 성능이 높은 것뿐 아니라, feature importance를 통해 어떤 변수들이 이탈 예측에 중요한 역할을 했는지도 확인할 수 있다.
실제 노트북에서 importance 상위 변수로는 다음과 같은 항목들이 확인되었다.

avg_days_btwn_sale_date
recency
frequency
avg_days_btwn_sales
days_first_to_last_purchase
year_plat_cntry_aov
total_sales
total_tickets

이는 모델이 단순히 고객 기본 정보만 보는 것이 아니라, 고객의 구매 간격, 최근성, 빈도, 누적 매출, 고객지원 활동 등 복합적인 행동 패턴을 학습했다는 점을 보여준다. 즉 XGBoost는 노트북 2 데이터의 핵심 구조를 잘 반영한 모델이라고 해석할 수 있다.

4) threshold 조정 가능성까지 포함해 가장 실무적인 후보였기 때문이다

노트북 2에서도 이탈 클래스(1)를 더 잘 잡는 것이 중요하다고 보았기 때문에, 단순 predict()뿐 아니라 predict_proba()를 이용한 threshold 조정 가능성도 고려할 수 있다. XGBoost는 이런 확률 기반 조정이 가능하고, validation을 통해 recall과 precision의 균형을 맞출 수 있다는 점에서 실무 적용 가능성도 높다.

6.6) CatBoost를 함께 실험한 이유와 최종 선택에서의 위치

CatBoost 역시 tabular 데이터에서 매우 강력한 boosting 모델로 잘 알려져 있다.
특히 범주형 변수 처리가 강점이 있기 때문에, 노트북 2처럼 플랫폼, 국가, 세그먼트 등 범주형 정보가 포함된 데이터에 적합한 후보였다.

실험 결과 CatBoost도 train에서 매우 높은 성능을 보였고 validation에서도 약 0.96 수준의 accuracy를 보여, XGBoost와 거의 비슷한 수준의 강력한 성능을 나타냈다. 즉 성능만 놓고 보면 CatBoost 역시 최종 후보로 충분한 가치를 가진다.

다만 최종적으로는 노트북의 흐름상 XGBoost를 중심 모델로 채택하였다. 그 이유는:

tabular 데이터에서 boosting 기반의 강한 비선형 모델이라는 점이 분명했고
validation 성능이 매우 우수했으며
feature importance를 통해 모델 해석이 가능했고
실험 과정에서 XGBoost가 가장 일관되게 “좋은 상태”를 보여준다고 판단했기 때문이다.

즉 CatBoost도 매우 우수한 후보였지만, 노트북에서는 최종적으로 XGBoost가 가장 설명력 있고 설득력 있는 모델로 선택된 것이다.

7) 종합 해석

노트북 2의 모델링 실험은 단순한 baseline에서 출발해, 더 복잡한 비선형 구조를 반영할 수 있는 앙상블 및 boosting 모델로 확장하는 흐름을 따랐다.

Logistic Regression은 선형 baseline으로 적절했으나, 데이터의 복합 구조를 완전히 설명하기에는 한계가 있었다.
Decision Tree는 규칙 기반 비선형 관계를 반영했지만, 단일 모델로서 표현력에 한계가 있었다.
Random Forest는 더 안정적인 비선형 앙상블을 제공했지만, boosting 모델만큼의 세밀한 성능 향상은 제한적이었다.
VotingClassifier는 여러 모델의 장점을 평균화하려는 시도였으며 유의미한 결과를 냈지만, 데이터 구조를 직접적으로 학습하는 고성능 boosting 계열보다는 한 단계 아래의 성격을 보였다.
XGBoost와 CatBoost는 가장 우수한 성능을 보였고, 그중에서도 XGBoost가 높은 validation 성능과 feature importance 해석 가능성 측면에서 최종 모델로 선택되었다.

즉 노트북 2의 모델 선택 과정은 데이터의 복잡도에 맞추어 단순 선형 모델에서 고성능 비선형 boosting 모델로 점진적으로 발전한 과정으로 볼 수 있으며, 최종적으로 XGBoost를 선택한 것은 데이터 구조와 평가 결과 모두를 고려한 합리적인 판단이라고 정리할 수 있다.
